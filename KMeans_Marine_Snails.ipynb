```python
X=np.array([3,5,8])
Y=np.array([7,12,15])
```

```python
beta1=np.sum((X-np.mean(X))*(Y-np.mean(Y)))/np.sum((X-np.mean(X))**2)
beta1
```

```python
beta0=np.mean(Y)-beta1*np.mean(X)
beta0
```

```python
df=pd.DataFrame({'X':[3,5,8],'Y':[7,12,15]})
x=df[['X']]
y=df['Y']
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x,y)
```

```python
df_scaled = df2.apply(zscore)
```

```python
df_scaled.head()
```

```python
model = KMeans(n_clusters = 3)
```

```python
model
```

```python
cluster_range = range( 1, 15 )
cluster_errors = []
for num_clusters in cluster_range:
  model = KMeans( num_clusters)
  model.fit(df_scaled)
  labels = model.labels_
  centroids = model.cluster_centers_
  cluster_errors.append( model.inertia_ )
```

```python
6799-5837
```

```python
clusters_df = pd.DataFrame( { "num_clusters":cluster_range, "cluster_errors": cluster_errors } )
clusters_df[0:15]
```

```python
6799-5837
```

The total sum of squared distances of every data point from respective centroid is also called inertia. Let us print the inertia value for all K values. That K at which the inertia stop to drop significantly (elbow method) will be the best K.

```python
# Elbow plot

plt.figure(figsize=(12,6))
plt.plot( clusters_df.num_clusters, clusters_df.cluster_errors, marker = "o" )
```

```python
kmeans = KMeans(n_clusters=3, n_init = 15, random_state=2)
```

```python
kmeans.fit(df_scaled)
```

```python
centroids=kmeans.cluster_centers_
centroids
```
